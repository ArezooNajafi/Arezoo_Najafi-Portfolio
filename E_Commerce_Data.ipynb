```python
!pip install pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/data.csv' , encoding='latin1',dtype={"CustomerID": "string"})

df.head(5)
```

```python
# Checking null values

print(df.isna().sum())
```

```python
con=df['CustomerID'].isna()
filter_df=df[con]
print(filter_df)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Sale']=df['Quantity']*df['UnitPrice']
df['Date'] = df['InvoiceDate'].dt.date
df['Hour'] = df['InvoiceDate'].dt.hour
df['Day'] = df['InvoiceDate'].dt.day
df['Month'] = df['InvoiceDate'].dt.month
df['Year'] = df['InvoiceDate'].dt.year
df.head(5)
```

```python
df.info()
```

```python
df['CustomerID'] = df['CustomerID'].astype(str)
df['Year'] = df['Year'].astype(str)
df['Month'] = df['Month'].astype(int)
df['Day'] = df['Day'].astype(int)
df['Hour'] = df['Hour'].astype(int)
df.info()
```

```python
# Group and sort by Sale
df_top_product_sale = df.groupby('Description')['Sale'].sum().sort_values(ascending=False).head(5)

# Group and sort by Quantity
df_top_product_quantity = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(5)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Top 5 Sale Products
df_top_product_sale.plot(kind='bar', ax=axes[0], title='Top 5 Gross Sale Product', color='royalblue')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Sale')

# Plot Top 5 Quantity Products
df_top_product_quantity.plot(kind='bar', ax=axes[1], title='Top 5 Gross Quantity Product', color='seagreen')
axes[1].set_xlabel('Product')
axes[1].set_ylabel('Quantity')

# Rotate x-axis labels
for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Net Sale(Removing refund and return (Q<0))

Net_df=df[df['Quantity']>0]
df_top_product_sale = Net_df.groupby('Description')['Sale'].sum().sort_values(ascending=False).head(5)

# Group and sort by Quantity
df_top_product_quantity = Net_df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(5)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Top 5 Sale Products
df_top_product_sale.plot(kind='bar', ax=axes[0], title='Top 5 Net Sale Product', color='royalblue')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Sale')

# Plot Top 5 Quantity Products
df_top_product_quantity.plot(kind='bar', ax=axes[1], title='Top 5 Net Quantity Product', color='seagreen')
axes[1].set_xlabel('Product')
axes[1].set_ylabel('Quantity')

# Rotate x-axis labels
for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()

```

```python
# Customers with highest Sale

df_top_customer_sale = Net_df.groupby('CustomerID')['Sale'].sum().sort_values(ascending=False).head(5)

# Group and sort by Quantity
df_top_customer_quantity = Net_df.groupby('CustomerID')['Quantity'].sum().sort_values(ascending=False).head(5)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Top 5 Sale Products
df_top_customer_quantity.plot(kind='bar', ax=axes[0], title='Top 5 Net Customer By Sale', color='royalblue')
axes[0].set_xlabel('Customer')
axes[0].set_ylabel('Sale')

# Plot Top 5 Quantity Products
df_top_customer_quantity.plot(kind='bar', ax=axes[1], title='Top 5 Net Customer By Quanity', color='seagreen')
axes[1].set_xlabel('Customer')
axes[1].set_ylabel('Quantity')

# Rotate x-axis labels
for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()

```

```python
# Evaluate refund
refund_df = df[df['Quantity'] < 0]

# Group by Description and calculate total refund amount (negative sales)
df_top_refund_amount = refund_df.groupby('Description')['Sale'].sum().abs().sort_values(ascending=False).head(5)

# Group by Description and calculate total refunded quantity (negative quantity)
df_top_refund_quantity = refund_df.groupby('Description')['Quantity'].sum().abs().sort_values(ascending=False).head(5)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Top 5 Sale Products (Refund Amounts)
df_top_refund_amount.plot(kind='bar', ax=axes[0], title='Top 5 Products with the Largest Refunds', color='royalblue')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Refund')

# Plot Top 5 Quantity Products (Refund Quantities)
df_top_refund_quantity.plot(kind='bar', ax=axes[1], title='Top 5 Products by Return Quantity', color='seagreen')
axes[1].set_xlabel('Product')
axes[1].set_ylabel('Quantity')

# Rotate x-axis labels
for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Customers with highest Refund

df_top_customer_refund_amount = refund_df.groupby('CustomerID')['Sale'].sum().abs().sort_values(ascending=False).head(5)

# Group and sort by Quantity
df_top_customer_quantity_refund = Net_df.groupby('CustomerID')['Quantity'].sum().abs().sort_values(ascending=False).head(5)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Top 5 Sale Products
df_top_customer_refund_amount.plot(kind='bar', ax=axes[0], title='Top 5 Customers with the Highest Refund Amount', color='royalblue')
axes[0].set_xlabel('Customer')
axes[0].set_ylabel('Sale')

# Plot Top 5 Quantity Products
df_top_customer_quantity_refund.plot(kind='bar', ax=axes[1], title='Top 5 Customers with the Highest Refund Quanity', color='seagreen')
axes[1].set_xlabel('Customer')
axes[1].set_ylabel('Quantity')

# Rotate x-axis labels
for ax in axes:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()

```

```python
import folium
import pandas as pd

# Load GeoJSON file
world_geojson = "https://raw.githubusercontent.com/johan/world.geo.json/master/countries.geo.json"

# Net Sale (Removing refund and return where Q < 0)
Net_df = df[df['Quantity'] > 0]
df_top_loc_sale = Net_df.groupby('Country')['Sale'].sum().reset_index()

# Convert country names to match GeoJSON format if needed
country_name_mapping = {
    "EIRE": "Ireland",
    "USA": "United States of America",
    "RSA": "South Africa"
}

df_top_loc_sale["Country"] = df_top_loc_sale["Country"].replace(country_name_mapping)

# Create a Folium map
m = folium.Map(location=[20, 0], zoom_start=2, tiles="cartodb positron")

# Add Choropleth Layer
folium.Choropleth(
    geo_data=world_geojson,
    name="Sales",
    data=df_top_loc_sale,
    columns=["Country", "Sale"],
    key_on="feature.properties.name",
    fill_color="BuPu",  # Blue-Purple for better contrast
    fill_opacity=0.9,   # Make colors more visible
    line_opacity=0.4,   # Adjust borders for clarity
    nan_fill_color="white",  # Countries with no data remain white
    nan_fill_opacity=0.6,
    legend_name="Total Sales ($)",
).add_to(m)

# Show map
```
